```python
%load_ext autoreload
%autoreload 2
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from postprocessing import retrieve_stack_info, plotstack
import astropy.units as u
```

```python
# filepath = "../testing/testing_data/consol_stacks_lrgs_zlim_elgclrgc_nu10gt2_e10gtpt3_o10_100pct_0.81_0.94_test.h5"
nreg=10
filepath = "/pscratch/sd/r/rhlozek/stacking/ACTxDESI/orient_by_desi_lrgc_100/stacks/enmap/actdesi_lrgs_p7_p75_jan29_%ireg/ACT_y_fid_consol_stacks_z0.70_0.75_lrgc_nocut_lrgc_nu10gt2_e10gtpt3_o10_100pct_0.7_1.0.h5"%nreg
```

## Retrieve the stack data in comoving Mpc

```python
MyStack_comov = retrieve_stack_info(filepath, format="constant_comoving", binsize=2, crop_center=0.0, r_m0_normalization=None, cap_filter=True)
```

## View the image

```python
img = plotstack(MyStack_comov.avg_img, MyStack_comov.radius) #, vmin=1.5e-6, vmax=2e-6)
```

## Plot the profiles

```python
for m in range(5):
    plt.errorbar(MyStack_comov.r, MyStack_comov.Cr_avg_profiles[m], yerr=MyStack_comov.Cr_errors_full[m], label=f"m={m}") # /np.sqrt(48)
plt.legend()
plt.title('unbinned')
#plt.show()
plt.savefig('mplot_nreg%i_unbinned.png'%nreg)
plt.clf()
for m in range(5):
    plt.errorbar(MyStack_comov.r_binned, MyStack_comov.Cr_avg_profiles_binned[m], yerr=MyStack_comov.Cr_errors_binned[m], label=f"m={m}")
plt.title('Nreg=%i'%nreg)

plt.legend()
plt.savefig('mplot_nreg%i.png'%nreg)
```
