```python
%load_ext autoreload
%autoreload 2
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from postprocessing import retrieve_stack_info
import astropy.units as u
```

```python
filepath = "../testing/testing_data/consol_stacks_lrgs_zlim_elgclrgc_nu10gt2_e10gtpt3_o10_100pct_0.81_0.94.h5"
```

## Retrieve the stack data as an object

```python
MyStack = retrieve_stack_info(filepath, format="constant_comoving", binsize=2, crop_center=2.0, r_m0_normalization=8)
```

## View the image

```python
plt.imshow(MyStack.avg_img, origin='lower')
```

## Plot the profiles

```python
for m in range(5):
    plt.errorbar(MyStack.r, MyStack.Cr_avg_profiles[m], yerr=MyStack.Cr_errors_full[m], label=f"m={m}") # /np.sqrt(48)
plt.legend()
plt.show()
plt.clf()
for m in range(5):
    plt.errorbar(MyStack.r_binned, MyStack.Cr_avg_profiles_binned[m], yerr=MyStack.Cr_errors_binned[m], label=f"m={m}")
plt.legend()
```
