```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
exstack = np.loadtxt("../testing/testing_data/example_stack.txt")
```

```python

PilImg = Image.fromarray(exstack)

imgsize = exstack.shape[0]
resized_sidelen = imgsize*.6
adjust = (imgsize-resized_sidelen)/2.
CroppedImg = PilImg.crop((adjust, adjust, imgsize-adjust, imgsize-adjust))
RS = CroppedImg.resize((imgsize, imgsize))
```

```python
plt.imshow(np.array(PilImg))
print(np.array(PilImg).shape)
plt.show()
plt.clf()
plt.imshow(np.array(CroppedImg))
print(np.array(CroppedImg).shape)
plt.show()
plt.clf()
plt.imshow(np.array(RS))
print(np.array(RS).shape)
```
